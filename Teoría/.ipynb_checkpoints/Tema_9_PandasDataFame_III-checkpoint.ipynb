# DATAFRAMES III

```python
import pandas as pd
```

```python
df = pd.read_csv("Iris.csv")
df
```

## Visualización de frecuencias usando CROSSTAB

```python
pd.crosstab(df.SepalLengthCm, df.Species)
```

```python
pd.crosstab(df.SepalLengthCm, df.Species).plot(kind="bar")
```

```python
features = ["SepalLengthCm", "SepalWidthCm"]
for feature in features:
    pd.crosstab(df[feature], df["Species"]).plot(kind="bar")
```

## Uso del filtrado usando ".LOC"

```python
df
```

```python
# Mostrar datos desde la fila 4 hasta la fila 20 incluida y todas las columnas
df.loc[4:20, :]
```

```python
df.loc[5]
```

```python
# Mostrar por index concreto
df.loc[df.index[5], :]
```

```python
# Mostrar columnas concretas
df.loc[4:20, "Species"]
```

```python
df.loc[:, "Species"]
```

```python
df.loc[:, ["PetalLengthCm", "Species"]]
```

```python
# Filtrar una especie: ejemplo mostrar todos los valores de "Iris-versicolor"
df.loc[df["Species"] == "Iris-versicolor"]
```

```python
df.loc[lambda df: df["Species"] == "Iris-setosa"]
```
